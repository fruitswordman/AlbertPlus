```python
from datetime import datetime
import pytz

def convert_to_cst(timestamp):
    utc_time = datetime.utcfromtimestamp(timestamp)
    utc_time = utc_time.replace(tzinfo=pytz.utc)
    cst_time = utc_time.astimezone(pytz.timezone('Asia/Shanghai'))
    return cst_time

def extract_day_time(timestamp):
    cst_time = convert_to_cst(timestamp)
    day = cst_time.strftime('%Y-%m-%d')
    time = cst_time.strftime('%H:%M:%S')
    return day, time

# Example usage:
timestamp = 1698515700
day, time = extract_day_time(timestamp)
print(f"Day: {day}, Time: {time}")
```
